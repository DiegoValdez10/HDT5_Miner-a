```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('student.csv', header=None)
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
df.head()
```

```python
print("La data tiene: ", df.shape[0], " filas y ", df.shape[1], " columnas")
```

```python
col_names = ['Id','Student_Age','Sex','High_School_Type','Scholarship','Additional_Work','Sports_activity',
             'Transportation','Weekly_Study_Hours','Attendance','Reading','Notes','Listening_in_Class',
             'Project_work','Grade']

df.columns = col_names


df.columns
```

```python
df.info()
```

```python
categoricas = []

for columna in df.columns:
    if df[columna].dtype=='O':
        categoricas.append(columna)
        
print("Hay ", len(categoricas), " variables.")


print("Las variables categoricas son: ")
print("*"*100)
for var in categoricas:
    print(var)
```

```python
df[categoricas].head()
```

```python
df[categoricas].isnull().sum()
```

```python
for var in categoricas:
    print(df[var].value_counts())
    print("*"* 35)
```

```python
#Vemos la frecuencia de los valores en porcentajes

for var in categoricas:
    print(df[var].value_counts()/(len(df)))
    print('*'*50)
```

```python
#Vemos la cardinalidad 

for var in categoricas:
    print(var, ' tiene ', len(df[var].unique()), ' etiquetas')
    
#Ojo con native_country
```

```python
df['Id'] = pd.to_numeric(df['Id'], errors='coerce')
df['Student_Age'] = pd.to_numeric(df['Student_Age'], errors='coerce')
```

```python
numericas = []

for columna in df.columns:
    if df[columna].dtype!='O':
        numericas.append(columna)
        
print("Hay ", len(numericas), " variables.")


print("Las variables numericas son: ")
print("*"*100)
for var in numericas:
    print(var)
```

```python
df[numericas].head()
```

```python
df[numericas].isnull().sum()
```

```python
df.dropna(subset=numericas, inplace=True)
```

```python
X = df.drop(['High_School_Type'], axis = 1)
y = df['High_School_Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```
